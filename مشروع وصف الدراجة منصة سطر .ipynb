```python
class Bike :
    def __init__ (self,description,sale_price,cost, condition) :
        self.description = description
        self.sale_price = sale_price
        self.cost = cost
        self.condition = condition
    
    def create_bike(self):
        return {
            'description': self.description,
            'cost': self.cost,
            'sale_price': self.sale_price,
            'condition': self.condition,
            'sold': False
        }
    def update_sale_price(self,create_bike, sale_price):
        if create_bike['sold'] == True:
            print('Action not allowed, Bike has already been sold')
        else:
            create_bike['sale_price'] = self.sale_price 
    def sell(self,create_bike):
        create_bike['sold'] = True
```

```python
b = Bike('Univega Alpina, orange', cost=100, sale_price=500, condition=0.5)
```

```python
b.create_bike()
```

```python
b.update_sale_price(b.create_bike(),350)
```

```python
b.sell(b.create_bike())
```
